```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URL="https://www.reliancedigital.in/air-conditioners/c/S101510"
```

```python
# Header for requests
HEADERS=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36','Accept-Langugae':'en-US, en;q=0.5'})
```

```python
#http requests
webpage=requests.get(URL,headers=HEADERS)
webpage
```

```python
type(webpage.content)
```

```python
#Soup containing  all data
Soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
Soup
```

```python
#fetch the link as list of tag objects
links=Soup.find_all('a',attrs={'class':"nav__description__hoverULine"})
```

```python
links
```

```python
link =links[0].get('href')
```

```python
product_link="https://www.reliancedigital.in/"+link
```
